## In this project we are going to explore kaggle titanic dataset and learn how to do logictic regression.

Importing the libraries needed

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic_df = pd.read_csv("train.csv")
```

```python
titanic_df.head(10)
```

```python
titanic_df.shape
```

```python
titanic_df.describe()
```

```python
titanic_df["Age"].fillna(titanic_df["Age"].mean(), inplace = True)
```

```python
bin_edges = [0,20,40,60,80]
bin_names = ["0-20","21-40","41-60","61-80"]
titanic_df["age_group"] = pd.cut(titanic_df["Age"], bin_edges,labels= bin_names)
```

Columns with no data

```python
titanic_df.columns[titanic_df.isnull().any()]
```

no of people with no cabin details is too high we could either drop this column or find something to replace with none values like the alphabet of cabin. Here we will be deleting the column 

```python
titanic_df.drop(columns = ["Cabin","PassengerId","Name","Embarked"], axis =1, inplace = True )
titanic_df.head()
```

```python
titanic_df["intercept"] = 1
```

```python
titanic_df["Pclass"].nunique()
```

```python
titanic_df[["upper","middle","lower"]] = pd.get_dummies(titanic_df["Pclass"])
```

```python
titanic_df[["female","male"]] = pd.get_dummies(titanic_df["Sex"])
```

```python
age_group = titanic_df["age_group"].unique()
```

```python
titanic_df[["0-20","21-40","41-60","61-80"]] = pd.get_dummies(titanic_df["age_group"])
```

```python
titanic_df.head()
```

```python
import statsmodels.api as sm
```

```python
logit = sm.Logit(titanic_df["Survived"],titanic_df[["intercept","SibSp","Parch","Fare","upper","middle","male","0-20","21-40","41-60"]])
results = logit.fit()
results.summary2()
```

Here features with low p values means that they have more effect on the output

```python
df_test = pd.read_csv("test1.csv")
```

```python
df_test.head()
```

```python
df_test["age_group"] = pd.cut(df_test["Age"], bin_edges,labels= bin_names)
```

```python
df_test[["upper","middle","lower"]] = pd.get_dummies(df_test["Pclass"])
df_test[["female","male"]] = pd.get_dummies(df_test["Sex"])
df_test[["0-20","21-40","41-60","61-80"]] = pd.get_dummies(df_test["age_group"])
```

```python
df_test["intercept"] = 1
```

```python
prediction = results.predict(df_test[["intercept","SibSp","Parch","Fare","upper","middle","male","0-20","21-40","41-60"]])
print(prediction)
```

```python
predicted_choice = (prediction > 0.5).astype(int)
```

```python
print(predicted_choice)
```

```python
df = pd.DataFrame(data=predicted_choice, columns =["Survived"])
df.head()
```

```python
df["PassengerId"] = df_test["PassengerId"]
```

```python
df.head()
```

```python
df = df[["PassengerId","Survived"]]
```
